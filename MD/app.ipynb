### Objetivo:

```python
Realizar una herramienta de automatización para la instalación de programas en linux
```

### Alcance:

```python
Tener las dependencias obligatorias, verificar que no haya conflictos entre los paquetes 
y que se instale ordenadamente
```

### Especificaciones léxicas:

```python
Palabras reservadas: DEPENDE, REQUIERE, OPCIONAL, CONFLICTO, INSTALAR
Simbolos:{} :
Operadores: =, ==, >
Variables: 'Entorno', 'Version' , 'Paquete'  
Identificadores:'python', 'libssl', 'nodejs'
Literales: números, strings
```

### Especificaciones sintácticas:

*Estructura de Asignación de Variables
   Debe comenzar con la palabra reservada VAR.

  Seguido de un identificador de variable (entorno, paquete, version).

  Debe incluir el operador de asignación =.

  Termina con un literal (String entre comillas o Número) o un identificador.

*Estructura de Definición de Paquete
   Inicia con la palabra reservada PAQUETE seguida del nombre (identificador).

   El cuerpo del paquete debe estar delimitado por llaves { y }.

*Dentro de las llaves:
   VERSION: Seguido de un número.

   DEPENDE: Seguido del identificador de otro paquete.

   CONFLICTO: Seguido del identificador de un paquete incompatible.

*Estructura de Control (Condicionales)

  Condición: Una comparación entre una variable y un literal usando el operador ==.

  Acción: La palabra ENTONCES seguida de la instrucción INSTALAR y el nombre del paquete.

### Especificaciones semánticas:

Resolución de Dependencias: Antes de instalar un paquete A, el sistema debe verificar si tiene cláusulas DEPENDE. 
Si las tiene, los paquetes listados deben instalarse primero.
Recursividad: Si el paquete A depende de B, y B depende de C, el orden de instalación debe ser C -> B -> A.
Error de Validación: Si el usuario intenta instalar un paquete que entra en conflicto con uno ya instalado, 
la herramienta debe detener la ejecución y lanzar un error de "Incompatibilidad".

```python
VAR entorno = "desarrollo"
VAR paquete = nodejs
VAR version = 18.0

PAQUETE nodejs {
    VERSION 18.0
    DEPENDE libssl
    DEPENDE python
}

PAQUETE python {
    VERSION 3.10
    DEPENDE libssl
    CONFLICTO python2
}

PAQUETE python2 {
    VERSION 2.7
}

PAQUETE libssl {
    VERSION 1.1
}

PAQUETE dev_tools {
    VERSION 1.0
}

SI entorno == "produccion" ENTONCES INSTALAR nodejs

SI entorno == "desarrollo" ENTONCES INSTALAR dev_tools
```
