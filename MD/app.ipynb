## Lenguaje a crear (completar)

### Objetivo:

```python
Realizar una herramienta de automatización para la instalación de programas en linux
```

### Alcance:

```python
Tener las dependencias obligatorias, verificar que no haya conflictos entre los paquetes 
y que se instale ordenadamente
```

### Especificaciones léxicas:

```python
Palabras reservadas: VAR, PAQUETE, VERSION, DEPENDE, REQUIERE, OPCIONAL, CONFLICTO, SI, ENTONCES, INSTALAR
Simbolos: { }
Operadores: =, ==, >
Variables: entorno, version, paquete
Identificadores de paquetes: python, python2, libssl, nodejs, dev_tools
Literales: numeros y strings
```

### Especificaciones sintácticas:

*Asignacion de variables
  Forma: VAR identificador = valor
  El valor puede ser un string entre comillas, un numero o un identificador.
  Ejemplo: VAR entorno = "desarrollo"

*Definicion de paquetes
  Forma: PAQUETE identificador { cuerpo }
  El cuerpo del paquete debe estar delimitado por llaves { y }.
  Ejemplo: PAQUETE nodejs { VERSION 18.0 DEPENDE libssl }

*Sentencias dentro de un paquete
  VERSION numero: indica la version del paquete.
  DEPENDE identificador: declara una dependencia obligatoria.
  REQUIERE identificador: funciona como dependencia obligatoria.
  OPCIONAL identificador: declara una dependencia opcional.
  CONFLICTO identificador: declara un paquete incompatible.

*Condicional de instalacion
  Forma: SI variable operador valor ENTONCES INSTALAR paquete
  Operadores permitidos: == y >.
  Ejemplo: SI entorno == "desarrollo" ENTONCES INSTALAR dev_tools


### Especificaciones semánticas:

Validacion de paquetes: Todo paquete declarado con PAQUETE debe tener una VERSION.<br>
Validacion de referencias: Las dependencias, opcionales y conflictos deben apuntar a paquetes existentes.<br>
Validacion de condiciones: La variable usada en SI debe estar declarada con VAR y el paquete a instalar debe existir.<br>
Resolucion de dependencias: Antes de instalar un paquete A, se instalan primero sus dependencias.
Si A depende de B y B depende de C, el codigo generado instala en orden C -> B -> A.<br>
Validacion de conflictos: Antes de instalar un paquete se revisa si entra en conflicto con algun paquete ya instalado.<br>
Error semantico en ejecucion: Si hay incompatibilidad, la ejecucion se detiene con un error de Incompatibilidad.<br>
